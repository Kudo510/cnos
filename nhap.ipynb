```python
from PIL import Image

scene_id = 1
for frame_id in range(56):
    # Paths to the images
    cnos_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_cnos.png"
    foundpose_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_foundpose.png"
    cnos_foundpose_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_cnos_foundpose.png"
    groundtruth_result = cnos_foundpose_result
    
    # Open the images
    cnos_image = Image.open(cnos_result)
    foundpose_image = Image.open(foundpose_result)
    cnos_foundpose_image = Image.open(cnos_foundpose_result)
    groundtruth_image = Image.open(groundtruth_result)
    
    # Get the dimensions of the images (assuming they are all the same size)
    width, height = cnos_image.size
    
    # Crop the right half of each image
    right_box = (width // 2, 0, width, height)
    cnos_image_cropped = cnos_image.crop(right_box)
    foundpose_image_cropped = foundpose_image.crop(right_box)
    cnos_foundpose_image_cropped = cnos_foundpose_image.crop(right_box)
    
    # Resize the groundtruth image to match the dimensions of the cropped right half
    groundtruth_image_resized = groundtruth_image.resize((width // 2, height))
    
    # Combine the groundtruth image and the right half of the cnos image side by side
    combined_image = Image.new('RGB', (width, height))
    combined_image.paste(groundtruth_image_resized, (0, 0))
    combined_image.paste(cnos_image_cropped, (width // 2, 0))
    
    # Create a new image with enough space to hold all four images in a 2x2 grid
    stacked_image = Image.new('RGB', (width, height * 2))
    
    # Paste the combined image at the top-left
    stacked_image.paste(combined_image, (0, 0))
    
    # Paste the foundpose_image_cropped at the top-right
    stacked_image.paste(foundpose_image_cropped, (width // 2, 0))
    
    # Paste the cnos_foundpose_image_cropped at the bottom-left
    stacked_image.paste(cnos_foundpose_image_cropped, (0, height))
    
    # Paste the groundtruth_image_resized again at the bottom-right
    stacked_image.paste(groundtruth_image_resized, (width // 2, height))
    
    # Save the stacked image
    output_path = f"cnos_foundpose_analysis/output_images/{frame_id:06d}_{scene_id:06d}_stacked.png"
    stacked_image.save(output_path)

print("Stacking completed for all frames.")
```

```python
import os
from PIL import Image
import numpy as np
import glob

def occlude_image(image_path, direction):
    # Open the image
    img = Image.open(image_path)
    
    # Convert to numpy array
    img_array = np.array(img)
    
    # Get image dimensions
    height, width, _ = img_array.shape
    
    # Create a mask based on the direction
    if direction == 'right':
        mask = np.ones_like(img_array)
        mask[:, width//2:, :] = 0
        return Image.fromarray((img_array * mask).astype('uint8'), 'RGB')
    elif direction == 'left':
        mask = np.ones_like(img_array)
        mask[:, :width//2, :] = 0
        return Image.fromarray((img_array * mask).astype('uint8'), 'RGB')
    elif direction == 'bottom':
        mask = np.ones_like(img_array)
        mask[height//2:, :, :] = 0
        return Image.fromarray((img_array * mask).astype('uint8'), 'RGB')
    elif direction == 'upper':
        mask = np.ones_like(img_array)
        mask[:height//2, :, :] = 0
        return Image.fromarray((img_array * mask).astype('uint8'), 'RGB')
    elif direction == 'diagonal':
        # Create four masks for the four diagonal occlusions
        mask1 = np.ones_like(img_array)
        mask2 = np.ones_like(img_array)
        mask3 = np.ones_like(img_array)
        mask4 = np.ones_like(img_array)
        for i in range(height):
            for j in range(width):
                if i > j:
                    mask1[i, j, :] = 0  # Lower-left triangle
                if i < j:
                    mask2[i, j, :] = 0  # Upper-right triangle
                if i + j < width:
                    mask3[i, j, :] = 0  # Upper-left triangle
                if i + j >= width:
                    mask4[i, j, :] = 0  # Lower-right triangle
        
        # Apply the masks and create four images
        occluded_img1 = Image.fromarray((img_array * mask1).astype('uint8'), 'RGB')
        occluded_img2 = Image.fromarray((img_array * mask2).astype('uint8'), 'RGB')
        occluded_img3 = Image.fromarray((img_array * mask3).astype('uint8'), 'RGB')
        occluded_img4 = Image.fromarray((img_array * mask4).astype('uint8'), 'RGB')
        
        return occluded_img1, occluded_img2, occluded_img3, occluded_img4
    else:
        raise ValueError("Invalid direction. Choose from 'right', 'left', 'bottom', 'upper', or 'diagonal'")

def generate_all_occlusions(image_path, output_dir, image_id):
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # List of all directions
    directions = ['right', 'left', 'bottom', 'upper', 'diagonal']

    for direction in directions:
        result = occlude_image(image_path, direction)
        
        if direction == 'diagonal':
            # Save all four diagonal occlusions
            diagonal_names = ['lower_left', 'upper_right', 'upper_left', 'lower_right']
            for i, img in enumerate(result):
                output_path = os.path.join(output_dir, f'occluded_{direction}_{diagonal_names[i]}_{image_id}.png')
                img.save(output_path)
                print(f"Saved: {output_path}")
        else:
            # Save single occlusion for other directions
            output_path = os.path.join(output_dir, f'occluded_{direction}_{image_id}.png')
            result.save(output_path)
            print(f"Saved: {output_path}")

    print("All occlusions generated and saved.")

# Usage
image_paths = glob.glob("foundpose_analysis/icbin/templates/train_pbr_images_templates/obj_000001_original/*png")
output_directory = "foundpose_analysis/icbin/extended_templates/"
for image_path in image_paths:
    image_id = image_path.split('/')[-1].split('.')[0]
    generate_all_occlusions(image_path, output_directory,image_id)

# image_path = 'foundpose_analysis/icbin/templates/train_pbr_images_templates/obj_000001_original/000001.png'
# output_directory = 'xoa_out_dir'

# generate_all_occlusions(image_path, output_directory)
```

```python
from src.model.utils import convert_npz_to_json_2
import glob
import os
from functools import partial
import multiprocessing
import os.path as osp
import logging
from tqdm import tqdm
import time
import json

result_paths = sorted(
    glob.glob(
        osp.join("contrastive_learning/output_npz/000001*.npz",
        )
    )
)
result_paths = sorted(
    [path for path in result_paths if "runtime" not in path]
)
num_workers = 10
logging.info(f"Converting npz to json requires {num_workers} workers ...")
pool = multiprocessing.Pool(processes=num_workers)
convert_npz_to_json_with_idx = partial(
    convert_npz_to_json_2,
    list_npz_paths=result_paths,
)
detections = list(
    tqdm(
        pool.imap_unordered(
            convert_npz_to_json_with_idx, range(len(result_paths))
        ),
        total=len(result_paths),
        desc="Converting npz to json",
    )
)
formatted_detections = []
for detection in tqdm(detections, desc="Loading results ..."):
    formatted_detections.extend(detection)

detections_path = f"contrastive_learning/output_json/icbin_ob1.json"
with open(detections_path, "w") as f:
    json.dump(formatted_detections, f)
logging.info(f"Saved predictions to {detections_path}")
```
