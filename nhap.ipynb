```python
from PIL import Image

scene_id = 1
for frame_id in range(56):
    # Paths to the images
    cnos_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_cnos.png"
    foundpose_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_foundpose.png"
    cnos_foundpose_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_cnos_foundpose.png"
    groundtruth_result = cnos_foundpose_result
    
    # Open the images
    cnos_image = Image.open(cnos_result)
    foundpose_image = Image.open(foundpose_result)
    cnos_foundpose_image = Image.open(cnos_foundpose_result)
    groundtruth_image = Image.open(groundtruth_result)
    
    # Get the dimensions of the images (assuming they are all the same size)
    width, height = cnos_image.size
    
    # Crop the right half of each image
    right_box = (width // 2, 0, width, height)
    cnos_image_cropped = cnos_image.crop(right_box)
    foundpose_image_cropped = foundpose_image.crop(right_box)
    cnos_foundpose_image_cropped = cnos_foundpose_image.crop(right_box)
    
    # Resize the groundtruth image to match the dimensions of the cropped right half
    groundtruth_image_resized = groundtruth_image.resize((width // 2, height))
    
    # Combine the groundtruth image and the right half of the cnos image side by side
    combined_image = Image.new('RGB', (width, height))
    combined_image.paste(groundtruth_image_resized, (0, 0))
    combined_image.paste(cnos_image_cropped, (width // 2, 0))
    
    # Create a new image with enough space to hold all four images in a 2x2 grid
    stacked_image = Image.new('RGB', (width, height * 2))
    
    # Paste the combined image at the top-left
    stacked_image.paste(combined_image, (0, 0))
    
    # Paste the foundpose_image_cropped at the top-right
    stacked_image.paste(foundpose_image_cropped, (width // 2, 0))
    
    # Paste the cnos_foundpose_image_cropped at the bottom-left
    stacked_image.paste(cnos_foundpose_image_cropped, (0, height))
    
    # Paste the groundtruth_image_resized again at the bottom-right
    stacked_image.paste(groundtruth_image_resized, (width // 2, height))
    
    # Save the stacked image
    output_path = f"cnos_foundpose_analysis/output_images/{frame_id:06d}_{scene_id:06d}_stacked.png"
    stacked_image.save(output_path)

print("Stacking completed for all frames.")
```

```python
from PIL import Image, ImageDraw, ImageFont
import os

# Load a larger font for the text
font_size = 60  # Font size for the text
try:
    font = ImageFont.truetype("arial.ttf", font_size)  # Specify the path to a font file if needed
except IOError:
    font = ImageFont.load_default()
    font = ImageFont.load_default().font_variant(size=font_size)

scene_id = 1
for frame_id in range(56):
    # Paths to the images
    cnos_04 = f"cnos_foundpose_analysis/output_images_different_thresholds/{scene_id:06d}_{frame_id:06d}_cnos_04.png"
    cnos_05 = f"cnos_foundpose_analysis/icbin/output_images/{scene_id:06d}_{frame_id:06d}_cnos.png"
    foundpose_result = f"cnos_foundpose_analysis/icbin/output_images/{scene_id:06d}_{frame_id:06d}_foundpose.png"
    cnos_foundpose_result = f"cnos_foundpose_analysis/icbin/output_images/{scene_id:06d}_{frame_id:06d}_cnos_foundpose.png"
    contrastive_learning = f"contrastive_learning/output_images_self_rotate/{scene_id:06d}_{frame_id:06d}_contrastive.png"
    
    # Open the images
    cnos_04_image = Image.open(cnos_04)
    cnos_05_image = Image.open(cnos_05)
    foundpose_image = Image.open(foundpose_result)
    cnos_foundpose_image = Image.open(cnos_foundpose_result)
    contrastive_learning_image = Image.open(contrastive_learning)
    
    # Get the dimensions of the images (assuming they are all the same size)
    width, height = cnos_04_image.size
    
    # Create an ImageDraw object for each image to add text
    draw_cnos_04 = ImageDraw.Draw(cnos_04_image)
    draw_cnos_05 = ImageDraw.Draw(cnos_05_image)
    draw_foundpose = ImageDraw.Draw(foundpose_image)
    draw_cnos_foundpose = ImageDraw.Draw(cnos_foundpose_image)
    draw_contrastive_learning = ImageDraw.Draw(contrastive_learning_image)
    
    # Define text positions and content
    text_position = (20, 20)  # Top-left corner with some padding
    text_color = "white"  # White text color
    
    # Add text to each image
    draw_cnos_04.text(text_position, "CNOS 04", fill=text_color, font=font)
    draw_cnos_05.text(text_position, "CNOS 05", fill=text_color, font=font)
    draw_foundpose.text(text_position, "Foundpose", fill=text_color, font=font)
    draw_cnos_foundpose.text(text_position, "CNOS + Foundpose", fill=text_color, font=font)
    draw_contrastive_learning.text(text_position, "Contrastive Learning", fill=text_color, font=font)
    
    # Create a new image with enough space to hold all five images vertically
    stacked_image = Image.new('RGB', (width, height * 5))
    
    # Paste the images into the new image
    stacked_image.paste(cnos_04_image, (0, 0))
    stacked_image.paste(cnos_05_image, (0, height))
    stacked_image.paste(foundpose_image, (0, height * 2))
    stacked_image.paste(cnos_foundpose_image, (0, height * 3))
    stacked_image.paste(contrastive_learning_image, (0, height * 4))
    
    # Save the stacked image
    output_dir = "cnos_foundpose_analysis/output_final_images"
    os.makedirs(output_dir, exist_ok=True)
    output_path = os.path.join(output_dir, f"{scene_id:06d}_{frame_id:06d}_stacked.png")
    stacked_image.save(output_path)

print("Stacking completed for all frames.")
```

```python
from PIL import Image, ImageDraw, ImageFont

# Define the path to a TrueType font file
font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf"  # Adjust this path as needed

# Load the font with a larger size
font_size = 80  # You can adjust this size if needed
font = ImageFont.truetype(font_path, font_size)

scene_id = 1
for frame_id in range(59):
    # Paths to the images
    cnos_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_cnos.png"
    foundpose_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_foundpose.png"
    cnos_foundpose_result = f"cnos_foundpose_analysis/output_images/{scene_id:06d}_{frame_id:06d}_cnos_foundpose.png"
    
    # Open the images
    cnos_image = Image.open(cnos_result)
    foundpose_image = Image.open(foundpose_result)
    cnos_foundpose_image = Image.open(cnos_foundpose_result)
    
    # Get the dimensions of the images (assuming they are all the same size)
    width, height = cnos_image.size
    
    # Create an ImageDraw object for each image to add text
    draw_cnos = ImageDraw.Draw(cnos_image)
    draw_foundpose = ImageDraw.Draw(foundpose_image)
    draw_cnos_foundpose = ImageDraw.Draw(cnos_foundpose_image)
    
    # Define text content
    text_color = "white"  # White text color
    
    # Calculate the position for the top-right corner
    cnos_text = "CNOS"
    foundpose_text = "Foundpose"
    cnos_foundpose_text = "CNOS + Foundpose"
    
    # Get text sizes
    cnos_text_size = draw_cnos.textsize(cnos_text, font=font)
    foundpose_text_size = draw_foundpose.textsize(foundpose_text, font=font)
    cnos_foundpose_text_size = draw_cnos_foundpose.textsize(cnos_foundpose_text, font=font)
    
    # Calculate the positions for the top-right corner
    cnos_position = (width - cnos_text_size[0] - 10, 10)
    foundpose_position = (width - foundpose_text_size[0] - 10, 10)
    cnos_foundpose_position = (width - cnos_foundpose_text_size[0] - 10, 10)
    
    # Add text to each image at the calculated positions
    draw_cnos.text(cnos_position, cnos_text, fill=text_color, font=font)
    draw_foundpose.text(foundpose_position, foundpose_text, fill=text_color, font=font)
    draw_cnos_foundpose.text(cnos_foundpose_position, cnos_foundpose_text, fill=text_color, font=font)
    
    # Create a new image with enough space to hold all three images vertically
    stacked_image = Image.new('RGB', (width, height * 3))
    
    # Paste the images into the new image
    stacked_image.paste(cnos_image, (0, 0))
    stacked_image.paste(foundpose_image, (0, height))
    stacked_image.paste(cnos_foundpose_image, (0, height * 2))
    
    # Save the stacked image
    output_path = f"cnos_foundpose_analysis/output_final_images/{frame_id:06d}_{scene_id:06d}_stacked.png"
    stacked_image.save(output_path)

print("Stacking completed for all frames.")
```

```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
from PIL import Image
import random
random.seed(10)

class MultiRotate(object):
    def __init__(self, angles):
        self.angles = angles

    def __call__(self, img):
        angle = random.choice(self.angles)
        return transforms.functional.rotate(img, angle)
    
# Load an image
image_path = 'cnos_analysis/crop_proposals/crop1.png'
img = Image.open(image_path)

# Define the transformations
# transform = transforms.Compose([
#     transforms.RandomHorizontalFlip(),
#     transforms.RandomRotation(30),
#     transforms.ColorJitter(brightness=0.5, contrast=0.5),
#     transforms.RandomResizedCrop(size=(256, 256)),
# ])

rotation_angles = [30, 60, 90, 120, 150]

transform = transforms.Compose(
    [
        # transforms.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
        # transforms.GaussianBlur(kernel_size=(5, 5), sigma=sigma_range), 
        # transforms.ColorJitter(brightness=(1.1, 1.2)),
        # transforms.ColorJitter(brightness=(0.7, 0.9)),
        # transforms.RandomAutocontrast(p=1)
        transforms.RandomHorizontalFlip(),  # Random horizontal flip
        transforms.RandomVerticalFlip(),  # Random horizontal flip
        MultiRotate(rotation_angles),  # Custom MultiRotate transformation

    ]
)

# Apply the transformation multiple times and display the results
fig, axes = plt.subplots(1, 10, figsize=(15, 5))
for i in range(10):
    augmented_img = transform(img)
    axes[i].imshow(augmented_img)
    axes[i].axis('off')

plt.show()
```
